### Esercizio 2.5 — Groupby e aggregazione

1. Raggruppa per `Job_Title` e calcola la media di `Salary_Min` per ciascun gruppo.
2. Raggruppa per `Job_Title` e calcola la mediana di `Salary_Min`.
3. Usa `agg()` per ottenere, in un'unica chiamata, il minimo, il massimo e la mediana di `Salary_Min` per ciascun `Job_Title`.
4. (Bonus) Raggruppa per `Location` invece che per `Job_Title` e calcola la media di `Salary_Max`.


```python
import pandas as pd
df = pd.read_csv("job_offers.csv")

# 1. Media Salary_Min per Job_Title
df.groupby("Job_Title")["Salary_Min"].mean()

# 2. Mediana Salary_Min per Job_Title
df.groupby("Job_Title")["Salary_Min"].median()

# 3. agg con min, max, median
df.groupby("Job_Title")["Salary_Min"].agg(["min","max","median"])

# 4. Bonus: media Salary_Max per Location
df.groupby("Location")["Salary_Max"].mean()
```

### Esercizio 2.6 (bonus) — Filtri combinati

1. Filtra le offerte di lavoro che hanno `Job_Title` uguale a `'Data Scientist'` E `Salary_Min` superiore a 45000.
2. Filtra le offerte che hanno `Company` non mancante E sono basate a `'Remote'` come location.
3. Quante offerte di lavoro nel dataset hanno sia `Job_Title` uguale a `'ML Engineer'` sia provengono da `'Milano, IT'`?


```python
import pandas as pd
df = pd.read_csv("job_offers.csv")

# 1. Data Scientist con Salary_Min > 45000
df[(df["Job_Title"] == "Data Scientist") & (df["Salary_Min"] > 45000)]

# 2. Company non NA e Location == Remote
df[(df["Company"].isna() != True) & (df["Location"] == "Remote")]

#alternativa più pulita ~ è il not , volendo si potrebbe usare anche .notna()
#df[(~df["Company"].isna()) & (df["Location"] == "Remote")]


# 3. ML Engineer a Milano
df[(df["Job_Title"] == "ML Engineer") & (df["Location"] == "Milano, IT")]
```
